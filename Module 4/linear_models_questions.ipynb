```python
import pandas as pd

ames_housing = pd.read_csv("../datasets/ames_housing_no_missing.csv")
target_name = "SalePrice"
data = ames_housing.drop(columns=target_name)
target = ames_housing[target_name]
```

```python
numerical_features = [
    "LotFrontage", "LotArea", "MasVnrArea", "BsmtFinSF1", "BsmtFinSF2",
    "BsmtUnfSF", "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "LowQualFinSF",
    "GrLivArea", "BedroomAbvGr", "KitchenAbvGr", "TotRmsAbvGrd", "Fireplaces",
    "GarageCars", "GarageArea", "WoodDeckSF", "OpenPorchSF", "EnclosedPorch",
    "3SsnPorch", "ScreenPorch", "PoolArea", "MiscVal",
]

data_numerical = data[numerical_features]

data_numerical.head()
```

```python
#Q1

from sklearn.model_selection import cross_validate
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
import numpy as np

ridge = make_pipeline(
    StandardScaler(), Ridge(alpha=0)
)

cv_results = cross_validate(
    ridge,
    data_numerical,
    target,
    cv=10,
    scoring="neg_mean_squared_error",
    return_estimator=True,
)

max_coeff =  max([np.max(np.abs(est[-1].coef_)) for est in cv_results["estimator"]])
print(max_coeff)
```

```python
#Q2

ridge = make_pipeline(
    StandardScaler(), Ridge(alpha=1)
)

cv_results = cross_validate(
    ridge,
    data_numerical,
    target,
    cv=10,
    scoring="neg_mean_squared_error",
    return_estimator=True,
)

max_coeff =  max([np.max(np.abs(est[-1].coef_)) for est in cv_results["estimator"]])
print(max_coeff)
```

```python
# Q3
import matplotlib.pyplot as plt

coefs = [est[-1].coef_ for est in cv_results["estimator"]]
weights_ridge = pd.DataFrame(coefs, columns=data_numerical.columns)
weights_ridge.describe() #	2895.28764



# color = {"whiskers": "black", "medians": "black", "caps": "black"}
# fig, ax = plt.subplots(figsize=(8, 10))
# weights_ridge.plot.box(color=color, vert=False, ax=ax)
# _ = ax.set(title="Ridge regression weights")
```

```python
#Q4

data_without_GarageArea = data_numerical.drop(columns=["GarageArea"])

data_without_GarageArea

cv_results = cross_validate(
    ridge,
    data_without_GarageArea,
    target,
    cv=10,
    scoring="neg_mean_squared_error",
    return_estimator=True,
)

coefs = [est[-1].coef_ for est in cv_results["estimator"]]
weights_ridge = pd.DataFrame(coefs, columns=data_without_GarageArea.columns)
weights_ridge.describe() #std of GarageCars decreases because GarageArea and GarageCars are dependent
#Correlated features typically cause unstable estimation of the matching linear model coefficients, even with some level of regularization.


# color = {"whiskers": "black", "medians": "black", "caps": "black"}
# fig, ax = plt.subplots(figsize=(8, 10))
# weights_ridge.plot.box(color=color, vert=False, ax=ax)
# _ = ax.set(title="Ridge regression weights")
```

```python
#Q6 - 7

from sklearn.linear_model import RidgeCV

alphas = np.logspace(-3, 3, num=101)
ridge = make_pipeline(
    StandardScaler(),
    RidgeCV(alphas=alphas),
)

cv_results = cross_validate(
    ridge,
    data_without_GarageArea,
    target,
    cv=10,
    scoring="neg_mean_squared_error",
    return_estimator=True,
)

# print(len(cv_results["estimator"])) # 10

coefs = [est[-1].coef_ for est in cv_results["estimator"]]
weights_ridge = pd.DataFrame(coefs, columns=data_without_GarageArea.columns)
weights_ridge.describe() #std of GarageCars decreases because GarageArea and GarageCars are dependent

alphas_est = [est[-1].alpha_ for est in cv_results["estimator"]]
alphas_ridge = pd.DataFrame(alphas_est)
print(alphas_ridge)
alphas_ridge.describe()
# print(vars(cv_results["estimator"][0][-1]))
```

```python
#Q8
from sklearn.compose import make_column_selector as selector
from sklearn.preprocessing import OneHotEncoder    
from sklearn.compose import ColumnTransformer

# new_data = data.drop(columns="GarageArea")

numerical_columns_selector = selector(dtype_exclude=object)
categorical_columns_selector = selector(dtype_include=object)
numerical_columns = numerical_columns_selector(data)
categorical_columns = categorical_columns_selector(data)


preprocessor = ColumnTransformer(
    [("categorical", OneHotEncoder(handle_unknown="ignore"), categorical_columns),
     ("numerical", StandardScaler(), numerical_columns)]
)


alphas = np.logspace(-3, 3, num=101)
ridge = make_pipeline(
    preprocessor,
    RidgeCV(alphas=alphas),
)

cv_results_num_cat = cross_validate(
    ridge,
    data,
    target,
    cv=10,
    scoring="neg_mean_squared_error",
    return_estimator=True,
)

count = 0
for num_cat, num in zip(cv_results_num_cat["test_score"], cv_results["test_score"]):
    if (num > num_cat):
        count += 1

print(f'the numerical model is better in {count} situations')
```

```python
from sklearn.kernel_approximation import Nystroem
from sklearn.preprocessing import SplineTransformer


preprocessor = ColumnTransformer(
    [("categorical", OneHotEncoder(handle_unknown="ignore"), categorical_columns),
     ("numerical",  SplineTransformer(), numerical_columns)]
)

ridge = make_pipeline(
    preprocessor,
    Nystroem(kernel="poly", degree=2, n_components=300),
    RidgeCV(alphas=alphas),
)

cv_results_new = cross_validate(
    ridge,
    data,
    target,
    cv=10,
    scoring="neg_mean_squared_error",
    return_estimator=True,
)

count = 0
for new_eng, num_cat in zip(cv_results_new["test_score"], cv_results_num_cat["test_score"]):
    if (new_eng > num_cat):
        count += 1

print(f'the new model is better in {count} situations')
```
